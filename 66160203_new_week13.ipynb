```python
#จงตรวจสอบว่ากราฟในข้อ 1. มีวงจรแฮมิลตันหรือไม่ โดยใช้ทฤษฎีของ Ore หากเป็นไปตามทฤษฎีของ
#Ore ให้แสดงข้อความ"This graph has a Hamilton circuit." หากไม่เป็นไปตามทฤษฎีของ Ore
#แสดงข้อความ "Not define." ออกทางหน้าจอimport networkx as nx
import networkx as nx
import matplotlib.pyplot as plt
def theorem_ore (G1):
  if G1.number_of_nodes() < 3 :
    return -1
  n = G1.number_of_nodes()
  for node_u in G1:
    neighbors_u= G1.neighbors(node_u)
    for node_v in G1:
      if node_v not in neighbors_u and node_u is not node_v:
        degree_of_u = G1.degree(node_u)
        degree_of_v = G1.degree(node_v)
        if (degree_of_u+degree_of_v) < n:
          return -1
  return 1
nx.draw(G1, with_labels = True)
plt.show()
if theorem_ore (G1) == 1 :
  print("This graph has a Hamilton circuit.")
else:
  print("Not define.")
```

```python
#จงเขียนโปรแกรมเพื่อหาวิถีถ่วงน้ำหนักที่มีน้ำหนักต่ำที่สุดจากจุด 0 ไปยังจุด 5 จากกราฟที่กำหนด
import networkx as nx
import matplotlib.pyplot as plt
G3 = nx.Graph()
G3.add_nodes_from(['0','1','2','3','4','5','6','7','8'])
G3.add_weighted_edges_from([('0','1',4),('0','6',7),('1','6',11),
('1','7',20),('1','2',9),('6','7',1),('2','3',6),('2','4',2),
('4','7',1),('4','3',10),('4','5',15),('4','8',5),('7','8',3),('3','5',5),('8','5',12)])
print('Shortest path is',nx.dijkstra_path(G3, source='0',target='5'))
print('Shortest path length is',nx.dijkstra_path_length (G3, source='0', target='5'))
pos = nx.spring_layout(G3)
nx.draw_networkx(G3, pos, with_labels=True, font_weight='bold',node_color='orange')
labels = nx.get_edge_attributes(G3, 'weight')
nx.draw_networkx_edge_labels(G3, pos, edge_labels=labels)
plt.show()
```
